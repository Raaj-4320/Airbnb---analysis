```python
df['host_response_rate'] = df['host_response_rate'].str.replace('%', '')
df['host_acceptance_rate'] = df['host_acceptance_rate'].str.replace('%', '')
```

```python
fill_with_zero = [
    'host_response_rate', 'host_acceptance_rate', 'accommodates', 'beds',
    'minimum_maximum_nights', 'maximum_maximum_nights', 'availability_30',
    'availability_60', 'availability_90', 'availability_365',
    'number_of_reviews', 'number_of_reviews_ltm', 'number_of_reviews_l30d',
    'review_scores_rating', 'review_scores_accuracy',
    'review_scores_cleanliness', 'review_scores_checkin',
    'review_scores_communication', 'review_scores_location',
    'review_scores_value', 'calculated_host_listings_count',
    'calculated_host_listings_count_entire_homes',
    'calculated_host_listings_count_private_rooms',
    'calculated_host_listings_count_shared_rooms', 'reviews_per_month'
]

df[fill_with_zero] = df[fill_with_zero].fillna(0)
```

```python
columns_to_integer = ['host_response_rate',
                      'host_acceptance_rate',
                      'accommodates',
                      'beds',
                      'minimum_maximum_nights',
                      'maximum_maximum_nights',
                      'availability_30',
                      'availability_60',
                      'availability_90',
                      'availability_365',
                      'number_of_reviews',
                      'number_of_reviews_ltm',
                      'number_of_reviews_l30d',
                      'review_scores_rating',
                      'review_scores_accuracy',
                      'review_scores_cleanliness',
                      'review_scores_checkin',
                      'review_scores_communication',
                      'review_scores_location',
                      'review_scores_value',
                      'calculated_host_listings_count',
                      'calculated_host_listings_count_entire_homes',
                      'calculated_host_listings_count_private_rooms',
                      'calculated_host_listings_count_shared_rooms',
                      'reviews_per_month']


df[columns_to_integer] = df[columns_to_integer].astype(int)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the size of the matplotlib figure
plt.figure(figsize=(10, 8))  # Adjust the width and height as needed

# Add title
plt.title("Correlation heatmap")

# Plot the heatmap
sns.heatmap(data=numerical_columns, annot=True)

# Set the horizontal label
plt.xlabel("Airbnb")

# Show the plot
plt.show()
```
